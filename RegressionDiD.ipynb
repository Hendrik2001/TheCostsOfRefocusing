Alternative approach, using log control variables instead of dummies

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels import PanelOLS

# Load and filter data
data = pd.read_csv('regression_df.csv')
filtered_data = data[(data['intheset'] == 1) & (data['mydate'] <= 617) & (data['firmscope'] < 50)]

# Prepare variables
y = filtered_data['ir']
continuous_vars = ['firmscope', 'aum_y', 'aum_firm', 'firm_age', 'age']
year_dummies = ['year_dum_1995', 'year_dum_1996', 'year_dum_1997', 'year_dum_1998', 'year_dum_1999', 'year_dum_2000',
    'year_dum_2001', 'year_dum_2002', 'year_dum_2003', 'year_dum_2004', 'year_dum_2005', 'year_dum_2006', 'year_dum_2007',
    'year_dum_2009', 'year_dum_2010', 'year_dum_2011']

# Log transform continuous variables (adding 1 to avoid log(0))
for var in continuous_vars:
    filtered_data[f'log_{var}'] = np.log1p(filtered_data[var])

X = filtered_data[['TREATED'] + [f'log_{var}' for var in continuous_vars] + year_dummies]

# Step 5: Add a constant to the model (for intercept)
X = sm.add_constant(X)

# Step 6: Run the OLS model with clustered standard errors
print("OLS Regression Results:")
try:
    ols_model = sm.OLS(y, X).fit(cov_type='cluster', cov_kwds={'groups': filtered_data['id']})
    print(ols_model.summary())
except Exception as e:
    print(f"An error occurred while fitting the OLS model: {str(e)}")

# Step 7: Prepare data for panel regressions
filtered_data['id'] = filtered_data['id'].astype('category')
filtered_data = filtered_data.set_index(['id', 'mydate'])

y = filtered_data['ir']
X = filtered_data[['TREATED'] + [f'log_{var}' for var in continuous_vars] + year_dummies]

# Step 8: Run the fixed effects model
print("\nFixed Effects Regression Results:")
try:
    fe_model = PanelOLS(y, X, entity_effects=True)
    fe_results = fe_model.fit(cov_type='clustered', cluster_entity=True)
    print(fe_results)
except Exception as e:
    print(f"An error occurred while fitting the fixed effects model: {str(e)}")

# Step 9: Run the fixed effects model with int1c
print("\nFixed Effects Regression Results with int1c:")
try:
    X_with_int1c = X.copy()
    X_with_int1c['int1c'] = filtered_data['int1c']
    fe_model_int1c = PanelOLS(y, X_with_int1c, entity_effects=True)
    fe_results_int1c = fe_model_int1c.fit(cov_type='clustered', cluster_entity=True)
    print(fe_results_int1c)
except Exception as e:
    print(f"An error occurred while fitting the fixed effects model with int1c: {str(e)}")

# Optionally, save results to a file
with open('regression_results_continuous.txt', 'w') as f:
    f.write("OLS Regression Results:\n")
    f.write(str(ols_model.summary()) + "\n\n")
    f.write("Fixed Effects Regression Results:\n")
    f.write(str(fe_results) + "\n\n")
    f.write("Fixed Effects Regression Results with int1c:\n")
    f.write(str(fe_results_int1c))

print("\nResults saved to 'regression_results_continuous.txt'")
```
